### Filling Data 

```python
train_data.info()
```

```python
train_data=train_data.drop(['Id','Alley','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
testID = test_data['Id']
test_data=test_data.drop(['Id','Alley','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
NumericalTrainData= train_data.select_dtypes(exclude=(object))
CategoricalTrainData = train_data.select_dtypes(include=(object))
NumericalTestData= test_data.select_dtypes(exclude=(object))
CategoricalTestData = test_data.select_dtypes(include=(object))
```

```python
NumNanTrainData = [ feature for feature in NumericalTrainData if train_data[feature].isnull().sum()>1 ]

for feature in NumNanTrainData:
  median_value = train_data[feature].median()
  train_data[feature].fillna(median_value,inplace=True)
```

```python
NumNanTestData = [ feature for feature in NumericalTestData if test_data[feature].isnull().sum()>=1 ]

for feature in NumNanTestData:
  median_value = test_data[feature].median()
  test_data[feature].fillna(median_value,inplace=True)
```

```python
train_data.info()
```

```python
train_data['MasVnrType']=train_data['MasVnrType'].fillna('None')
train_data['BsmtQual']=train_data['BsmtQual'].fillna('TA')
train_data['BsmtCond']=train_data['BsmtCond'].fillna('TA')
train_data['BsmtExposure']=train_data['BsmtExposure'].fillna('No')
train_data['BsmtFinType1']=train_data['BsmtFinType1'].fillna('Unf')
train_data['BsmtFinType2']=train_data['BsmtFinType2'].fillna('Unf')
train_data['Electrical']=train_data['Electrical'].fillna('SBrkr')
train_data['FireplaceQu']=train_data['FireplaceQu'].fillna('Gd')
train_data['GarageType']=train_data['GarageType'].fillna('Attchd')
train_data['GarageFinish']=train_data['GarageFinish'].fillna('Unf')
train_data['GarageQual']=train_data['GarageQual'].fillna('TA')
train_data['GarageCond']=train_data['GarageCond'].fillna('TA')
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
test_data['MSZoning']=test_data['MSZoning'].fillna('RL')
test_data['Utilities']=test_data['Utilities'].fillna('AllPub')
test_data['Exterior1st']=test_data['Exterior1st'].fillna('VinylSd')
test_data['Exterior2nd']=test_data['Exterior2nd'].fillna('VinylSd')
test_data['MasVnrType']=test_data['MasVnrType'].fillna('None')
test_data['BsmtQual']=test_data['BsmtQual'].fillna('TA')
test_data['BsmtCond']=test_data['BsmtCond'].fillna('TA')
test_data['BsmtExposure']=test_data['BsmtExposure'].fillna('No')
test_data['BsmtFinType1']=test_data['BsmtFinType1'].fillna('GLQ')
test_data['BsmtFinType2']=test_data['BsmtFinType2'].fillna('Unf')
test_data['KitchenQual']=test_data['KitchenQual'].fillna('TA')
test_data['Functional']=test_data['Functional'].fillna('Typ')
test_data['FireplaceQu']=test_data['FireplaceQu'].fillna('Gd')
test_data['GarageType']=test_data['GarageType'].fillna('Attchd')
test_data['GarageFinish']=test_data['GarageFinish'].fillna('Unf')
test_data['GarageQual']=test_data['GarageQual'].fillna('TA')
test_data['GarageCond']=test_data['GarageCond'].fillna('TA')
test_data['SaleType']=test_data['SaleType'].fillna('WD')
```

```python
test_data.info()
```

### Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for feature in CategoricalTrainData:
  train_data[feature]=lb.fit_transform(train_data[feature])
for feature in CategoricalTestData:
  test_data[feature]=lb.fit_transform(test_data[feature])
```

```python
train_data
```

### Feature Scaling

```python
feature_scale=[feature for feature in train_data.columns if feature not in ['SalePrice']]
feature_testScale=[feature for feature in test_data.columns]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scalerTest=MinMaxScaler()
scaler.fit_transform(train_data[feature_scale])
scalerTest.fit_transform(test_data[feature_testScale])
```

```python
data = pd.concat ( [train_data['SalePrice'].reset_index(drop=True),
                               pd.DataFrame(scaler.transform(train_data[feature_scale]),columns=feature_scale)],axis=1)
testdata= pd.DataFrame(scalerTest.transform(test_data[feature_testScale]),columns=feature_testScale)
```

```python
X= data.iloc[:,1:]
```

```python
y=data['SalePrice']
```

### Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### modeling

```python
from sklearn.ensemble import GradientBoostingRegressor

GBR=GradientBoostingRegressor(random_state=100)
GBR.fit(X_train,y_train)
```

```python
y_pred=GBR.predict(X_test)
print(GBR.score(X_train,y_train))
print(GBR.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)
```

### submission

```python
y_pred = GBR.predict(testdata)
```

```python
output = pd.DataFrame({'Id':testID,
                       'SalePrice': y_pred})
output.to_csv('submission.csv', index=False)
```
